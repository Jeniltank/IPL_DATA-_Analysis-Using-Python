```python
bowler_stats = pd.merge(dismissal_counts, match_bowler_perfomance, how='inner', on=['match_id', 'bowler'])

# Calculate Runs per Wicket
bowler_stats['Runs per Wicket'] = bowler_stats['Runs Conceded'] / bowler_stats['Wickets']

# Optional: handle division by zero if any bowler has 0 wickets (though unlikely after inner join)
bowler_stats['Runs per Wicket'].replace([float('inf'), -float('inf')], pd.NA, inplace=True)

# View result
bowler_stats.head()
```

```python
top_bowlers = bowler_stats.sort_values(by='Runs per Wicket').head(10)

# Set style
sns.set(style='whitegrid')

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_bowlers, x='Runs per Wicket', y='bowler', palette='viridis')

# Labels and title
plt.title('Top 10 Bowlers (Lowest Runs per Wicket)', fontsize=14, weight='bold')
plt.xlabel('Runs per Wicket', fontsize=12)
plt.ylabel('Bowler', fontsize=12)

# Add data labels
for index, row in top_bowlers.iterrows():
    plt.text(row['Runs per Wicket'] + 0.2, index, f"{row['Runs per Wicket']:.1f}", va='center')

plt.tight_layout()
plt.show()
```
